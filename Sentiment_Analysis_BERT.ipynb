```python
# Import necessary libraries

from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
# Istantiate Model
tokenizer = AutoTokenizer.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment-latest') # 0 Negative, 1 Neutral, 2 Positive
model = AutoModelForSequenceClassification.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment-latest')
```

```python
# testing encodes
def sentiment(comment):
  tokens = tokenizer.encode(comment, return_tensors='pt') #pt: PyTorch
  result = model(tokens) # Predict
  #Print the result
  if torch.argmax(result.logits) == 0:
    sen = 'Negative'
  elif torch.argmax(result.logits) == 1:
    sen = 'Neutral'
  else:
    sen = 'Positive'
  return sen
```

```python
comments = requests.get('https://dummyjson.com/comments?limit=300') # Dummy Data
data = comments.json()
values = []
for comment in data['comments']:
  values.append(comment['body'])
#soup = BeautifulSoup(tweet.content, 'html.parser')
#print(soup)
```

```python
df = pd.DataFrame(np.array(values), columns=['Comment'])
```

```python
df['Sentiment'] = df['Comment'].apply(sentiment)
```
